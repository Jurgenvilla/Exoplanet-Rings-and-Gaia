```python
# -*- coding: utf-8 -*-

import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
import sys
import struct
import math
import os

%matplotlib inline
```

```python
#Files with cut and without cut-off 

UCL        = Table.read('UCL_Table.csv', format = 'ascii', data_start = 0, fast_reader = True)
UCL_CutOff = Table.read('/disks/strw3/jvilla/Major_JorgeVilla/Scripts/6_Light_Curve/Cut-Off_Data/UCL/UCL_Table.csv', format = 'ascii', data_start = 0, fast_reader = True)
```

```python
Data_FITS   = []     #Fits
Data_Tables = []     #Tables
Name_1      = []     #Name without '_'
Name_2      = []     #Name with '_'

# for i in range (1,5):
for i in range (1, len(UCL)):
    #In order to use the same name as in the Table file and all the files
    A = UCL['sourceid'][i].encode('utf-8').strip()
    B = A.decode('utf8')
    C = UCL['sourceid'][i].replace(u'P J',u'P_J')

    #It Works for the FITS files 
    UCL_data = Table.read('/disks/strw3/jvilla/Major_JorgeVilla/Scripts/6_Light_Curve/UCL/'+B+'.fits', format = 'fits')
    Data_FITS.append(UCL_data)
    
    #It works for the tables 'lc.tbl'
    UCL_data_table = Table.read('/disks/strw3/jvilla/Major_JorgeVilla/Scripts/6_Light_Curve/UCL/'+C+'_lc.tbl', format = 'ascii')
    Data_Tables.append(UCL_data_table)
    
    Name_1.append(str(B))
    Name_2.append(str(C))
    
print('Done! for %i'%len(Name_1))    
```

```python
#Same process as performed above but only for the cut-off dataset

Data_FITS_CutOff   = []     #Fits
Data_Tables_CutOff = []     #Tables
Name_1_CutOff      = []     #Name without '_'
Name_2_CutOff      = []     #Name with '_'

# for i in range (1,5):
for i in range (1, len(UCL_CutOff)):
    #In order to use the same name as in the Table file and all the files
    A = UCL_CutOff['sourceid'][i].encode('utf-8').strip()
    B = A.decode('utf8')
    C = UCL_CutOff['sourceid'][i].replace(u'P J',u'P_J')

    #It Works for the FITS files 
    UCL_data = Table.read('/disks/strw3/jvilla/Major_JorgeVilla/Scripts/6_Light_Curve/Cut-Off_Data/UCL/'+B+'.fits', format = 'fits')
    Data_FITS_CutOff.append(UCL_data)
    
    #It works for the tables 'lc.tbl'
    UCL_data_table = Table.read('/disks/strw3/jvilla/Major_JorgeVilla/Scripts/6_Light_Curve/Cut-Off_Data/UCL/'+C+'_lc.tbl', format = 'ascii')
    Data_Tables_CutOff.append(UCL_data_table)
    
    Name_1_CutOff.append(str(B))
    Name_2_CutOff.append(str(C))
    
print('Done! for %i'%len(Name_1_CutOff))
```

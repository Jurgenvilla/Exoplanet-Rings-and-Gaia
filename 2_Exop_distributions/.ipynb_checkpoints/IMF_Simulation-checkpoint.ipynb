```python
import matplotlib.pyplot as plt
import numpy as np

import IMF_Generator

slopes   = [0.3, 1.3, 2.3, 3]
masslims = [0.01, 0.08, 0.5, 1, 2]
#slopes   = [0.3, 2.3, 5.3, -7]
#masslims = [1.0, 3.0, 7.0, 11.0, 15.]
#slopes = [2.35]
#masslims = [1,20]
myimf = IMF_Generator.MultiPartPowerLaw(np.array(slopes), np.array(masslims))
m     = np.linspace(masslims[0],masslims[-1],200)
lnpdf = myimf.pdf(m)
cdf   = myimf.cdf(m)
N     = 100000
ranmasses = myimf.rvs(N)

fig, ((axA, axB), (axC, axD)) = plt.subplots(figsize=(16,12), nrows=2, ncols=2)
axA.hist(ranmasses, bins='auto', normed=True, label='Simulated IMF')
axA.plot(m, np.exp(lnpdf), label='IMF')
axA.set_xlabel('$m$ [M$_\odot$]')
axA.set_ylabel('$p(m)$')
axA.legend()

axB.hist(ranmasses, bins='auto', normed=True, label='Simulated IMF')
axB.plot(m, np.exp(lnpdf), label='IMF')
axB.set_xlabel('$m$ [M$_\odot$]')
axB.set_ylabel('$p(m)$')
axB.set_yscale('log')
axB.set_xscale('log')
axB.legend()

c = np.linspace(1.0/N,1,N)
axC.plot(np.sort(ranmasses), c, label='Simulated cuml. MF')
axC.plot(m, cdf, label='Cumulative MF')
axC.set_xlabel('$m$ [M$_\odot$]')
axC.set_ylabel('$F(m)$')
axC.legend()

axD.plot(np.sort(ranmasses), c, label='Simulated cuml. MF')
axD.plot(m, cdf, label='Cumulative MF')
axD.set_xlabel('$m$ [M$_\odot$]')
axD.set_ylabel('$F(m)$')
axD.set_yscale('log')
axD.set_xscale('log')
axD.legend()

plt.show()
```
